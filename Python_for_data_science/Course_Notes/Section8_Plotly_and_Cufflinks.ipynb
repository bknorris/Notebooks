# Plotly and Cufflinks

* Plotly is an interactive visualization library
* Cufflinks connects plotly and pandas
* Install these libraries with pip install plotly and cufflinks

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns=['A','B','C','D'])
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'category':['A','B','C'],'values':[32,43,50]})
```

```python
df2
```

```python
df.plot.line()
```

```python
# Lets do this with plotly. Super cool! It's basically like MATLAB's standard plotting function!
df.iplot()
```

## Let's explore different plot types:

```python
# Scatter plot
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=8)
```

```python
# Bar plot
df2.iplot(kind='bar',x='category',y='values')
```

```python
df
```

```python
# Let's use an aggregate function on df to be able to create a barplot
df.sum().iplot(kind='bar')
```

```python
# Boxplots
df.iplot(kind='box')
```

```python
# 3D surface plot
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
# use a surface plot to create a 3d surface
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
# Histogram
df['A'].iplot(kind='hist',bins=25)
```

```python
# plot all the data
df.iplot(kind='hist',bins=25)
```

```python
# Spread, e.g., for stock data. Compare the "spread" between two stocks.
df[['A','B']].iplot(kind='spread')
```

```python
# Bubble plot
df.iplot(kind='bubble',x='A',y='B',size='D')
```

```python
# Scatter Matrix -- similar to seaborn's pairplot
df.scatter_matrix()
```
