# Data Visualization with Pandas

```python
import pandas as pd
import numpy as np
import seaborn as sns
path = '/mnt/c/Users/bknorris/Documents/Scripts/Python/Python_for_data_science/Py_DS_ML_Bootcamp/07-Pandas-Built-in-Data-Viz/'
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv(path+'df1',index_col=0)
```

```python
df2 = pd.read_csv(path+'df2')
```

```python
df1.head()
```

```python
df2.head()
```

```python
# A quick example of plotting with data frames in Pandas
df1['A'].hist()
```

```python
# Another way
df1['A'].plot(kind='hist')
```

```python
# A third way
df1['A'].plot.hist()
```

```python
# An area plot
df2.plot.area(alpha=0.5)
```

```python
# Bar plots. Good if your index is categorical.
df2.plot.bar(stacked=True)
```

```python
# Create a line plot for time series data
df1['A'].plot.line(x=df1.index,y='B')
```

```python
# Scatter plots
df1.plot.scatter(x='A',y='B')
```

```python
# Scatter plots, add some color via another column
df1.plot.scatter(x='A',y='B',c='C',cmap='magma')
```

```python
# Scatter plots, manipulate point scale by a fourth column
df1.plot.scatter(x='A',y='B',c='C',cmap='magma',s=df1['D']*50)
```

```python
# Box plots
df2.plot.box()
```

```python
# Plot some bivariate data
df = pd.DataFrame(np.random.randn(1000,2),columns=['A','B'])
```

```python
# Create a hex plot
df.plot.hexbin(x='A',y='B',gridsize=25,cmap='coolwarm')
```

```python
# KDE plots
df2['a'].plot.kde()
```
