# Matplotlib

* A popular plotting library
* A similar feel to MATLAB
* Works well with Pandas and Numpy arrays
* Seaborn is built off Matplotlib

Matplotlib allows you to create reproducible figures programmatically. Check out data visualization examples on http://matplotlib.org/

## Let's do some test plots!

```python
import matplotlib.pyplot as plt
```

```python
# Note: the line below is jupyter-notebook specific. It allows plots to be plotted "in line".
```

```python
%matplotlib inline
```

```python
# Let's create some dummy data
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# functional method for plotting
plt.plot(x,y,'-ob') # create a plot

# Basic labeling
plt.xlabel('Guns')
plt.ylabel('Butter')
plt.title('Simple Economics')
plt.show() # use this anywhere other than jupyter-notebooks
```

```python
# Create subplots
plt.subplot(1,2,1)
plt.plot(x,y,'--r')

plt.subplot(1,2,2)
plt.plot(y,x*2,'b')
plt.show()
```

```python
# Use object-oriented figure plotting... a better way to go.
# Create figure objects and call methods off of them.

fig1 = plt.figure() # create a figure object
axes = fig1.add_axes([0.1, 0.1, 0.8, 0.8]) #L B W H
axes.plot(x,y,'--b')
axes.set_xlabel('Guns')
axes.set_ylabel('Butter')
axes.set_title('Simple Economics')
```

```python
# Let's create a couple subplots with OOP
fig = plt.figure() # create a figure object

# define the location of the subplot axes
axes1 = fig.add_axes([0.1,0.1,0.35,0.8])
axes2 = fig.add_axes([0.55,0.1,0.35,0.8])
axes1.plot(x,y)
axes2.plot(y,x)
# add labels to each set of axes... etc...
```

```python
# more subplot routines. Calls fig.add_axes automatically!
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout() #automatic figure spacing
title = ['First plot','Second plot']

# axes is iterable
for id,current_ax in enumerate(axes):
    current_ax.plot(x,y)
    current_ax.set_title(title[id])
    
# Note I used enumerate here to iterate through title.
```

```python
# more subplot routines. Calls fig.add_axes automatically!
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout() #automatic figure spacing

# you can index it too,
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title('first plot')
axes[1].set_title('second plot')
```

## Figure size, Aspect ratio, and DPI

```python
# figure size
fig = plt.figure(figsize=(1,2),dpi=200) # width and height in inches; dpi is pixels per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# do this with subplots
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(4,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Legends

```python
figure = plt.figure(figsize=(2,2),dpi=200)
ax = figure.add_axes([0,0,1,1])

# This creates crappy "x carrot 2" label.
#ax.plot(x,x**2,'--b',label='x^2') # add plot labels at the end
#ax.plot(x,x**3,'-r',label='x^3')
#ax.legend() # create the legend

# The below example includes some fancier text
ax.plot(x,x**2,'--b',label=r'$x^2$') # add plot labels at the end
ax.plot(x,x**3,'-r',label=r'$x^3$')
ax.legend(loc='upper left') # create the legend, specify axes location

# NOTE: the above example with 'fancy' lettering only works if LaTeX is installed! 
# This might work out the gate with Spyder, maybe, but I had to 'pip install latex' here.
```

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) for details. 

## Changing plot colors


Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

### Colors with the color= parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Line width and line style

```python
fig, ax = plt.subplots()
ax.plot(x,y,color='purple',linewidth=0.5,linestyle='-.')
ax.plot(x,y+2,color='green',linewidth=1.5,linestyle='--')
ax.plot(x,y+4,color='red',linewidth=2.5,linestyle=':')
```

### Markers

```python
fig, ax = plt.subplots()
ax.plot(x,y,color='black',linewidth=0,linestyle='',
        marker='^',markersize=10,markerfacecolor='r')
```

### Control over axis appearance and plot range

```python
fig, ax = plt.subplots()
ax.plot(x,y,color='black',linewidth=0,linestyle='',
        marker='^',markersize=10,markerfacecolor='r')

# use set_xlim or set_ylim
ax.set_xlim([0,2])
ax.set_ylim([0,10])
```
